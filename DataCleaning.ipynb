#Feature Engineering

```python
# Additional variables for timestamp: year and year-month
X['year'] = X['timestamp'].apply(lambda x: int(x[0:4]))
X['year_mo'] = X['timestamp'].apply(lambda x: x[0:7])

# Residential & kitchen area to Total area ratio:
X['resident_to_total_ratio'] = X['life_sq']/X['full_sq']

# Average area per room:
X['avg_room_area'] = X['life_sq']/X['num_room']

# Extra area in the house and its ratio with respect to total area:
X['extra_area'] = X['full_sq'] - X['life_sq']
X['extra_area_ratio'] = X['extra_area']/X['full_sq']

# labor force percentage:
X['pct_labor_force'] = X['work_all']/X['raion_popul']

# Apartment floor relative to building height:
X['floor_rel_total'] = X['floor']/X['max_floor']

# Some additional binary variables:
X['metro_flag'] = np.where(X['raion_popul'] > 150000, 1, 0)
X['large_flag'] = np.where(X['max_floor'] >= 20, 1, 0)
X['small_flag'] = np.where(X['max_floor'] <= 20, 1, 0)

# Average building height for subarea:
sub_area = X.groupby('sub_area').agg({'max_floor':np.mean}).reset_index().rename(columns={'max_floor':'mean_bldg_height'})
X = pd.merge(X, sub_area, on = ['sub_area'], how = 'left')

# Monthly sales:
monthly_sales = X.groupby('year_mo').size().reset_index().rename(columns={0:'n_sales_month'})
X = pd.merge(X, monthly_sales, on = ['year_mo'], how = 'left')

# Average distance to Kremlin (subarea wise):
krem_dist = X.groupby('sub_area').agg({'kremlin_km':np.nanmean}).reset_index().rename(columns={'kremlin_km':'subarea_dist_to_kremlin'})
X = pd.merge(X, krem_dist, on = ['sub_area'], how = 'left')

# Count NaNs per row:
X['count_nan_per_row'] = X.isnull().sum(axis = 1)

# Apartment name:
X['apt_name'] = X['sub_area'] + X['metro_km_avto'].astype(str).apply(lambda x: x[0:5])
X['apt_name_yrmo'] = X['apt_name'] + X['year_mo']
```

# Imputing Missing Values

```python
m_values = (X.isna().sum()/X.shape[0])*100
m_values.sort_values(ascending=False, inplace=True)
m_values.head(40)
```

##Median Imputation (Missing value <=30%)

```python
print(len(m_values[(m_values>0) & (m_values<=30)].index))
imput_list = list(m_values[(m_values>0) & (m_values<=30)].index)
print(imput_list)
```

```python
imput_list.remove('product_type')
values = {col: X[col].median() for col in imput_list}
values['product_type'] = X['product_type'].mode()[0]
```

```python
X.fillna(value=values, inplace=True)
```

##Model based imputation for other missing values

```python
m_values[(m_values>30)].index
```

```python
from sklearn.impute import KNNImputer
from tqdm.notebook import tqdm_notebook

imputer = KNNImputer(n_neighbors=3)

col = list(m_values[(m_values>30)].index)

for i in tqdm_notebook(col):
    X[i] = imputer.fit_transform(X[[i]])
```

```python
from sklearn import preprocessing

numerical_cols = xTrain.select_dtypes(exclude=['object'])
categorical_cols = xTrain.select_dtypes(include=['object']).copy()

# Encoding for categorical columns:
for c in categorical_cols:
  le = preprocessing.LabelEncoder()
  le.fit(X_tr[c])

  X_val[c] = X_val[c].map(lambda s: '<unknown>' if s not in le.classes_ else s)
  le.classes_ = np.append(le.classes_, '<unknown>')

  X_tr[c] = le.transform(X_tr[c])
  X_val[c] = le.transform(X_val[c])

# Scaling for numercial columns:
for c in numerical_cols:
  min = X_tr[c].min()
  max = X_tr[c].max()

  X_tr[c] = (X_tr[c] - min)/(max-min)
  X_val[c] = (X_val[c] - min)/(max-min)
```
